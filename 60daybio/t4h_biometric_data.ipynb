File contains two dictionaries: ```Pheno```, ```NormalizedData``` (normalized w.r.t. first day of monitoring)
```Sid``` and ```Site``` **jointly** creates unique subject ID
```Sites```: identifiers of hospital (KGH, SPH, VGH)

````python
def read_from_pkl( file ):    
    from pickle import load    
    file = file.replace( '.pkl','')   
    file = open( file + '.pkl', 'rb')    
    try:
        dat = load( file )
    except:
        dat = load( file, encoding='latin1')
    return dat


#import myutils
#t4h= myutils.read_from_pkl('T4H_Normalized')
t4h = read_from_pkl('T4H_Normalized')
t4h.keys()
````
